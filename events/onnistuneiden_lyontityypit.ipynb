```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import cm
```

```python
csv_name = 'asd'
events_msu_20 = pd.read_csv('./events_2020_msu.csv'.format(csv_name), low_memory=False)
events_nsu_20 = pd.read_csv('./events_2020_nsu.csv'.format(csv_name), low_memory=False)
events_msu = pd.read_csv('./events_2021_msu.csv'.format(csv_name), low_memory=False)
events_nsu = pd.read_csv('./events_2021_nsu.csv'.format(csv_name), low_memory=False)
events_msu.head(5)
```

```python
ykskaks_vali = events_msu[events_msu['tilanne'] == '1']
ykskaks_vali_onnistuneet = ykskaks_vali[ykskaks_vali['tulos'] == 'Kärkilyönti']
c = ykskaks_vali_onnistuneet.groupby(by='tyyppi').count().sort_values(by='tulos')

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = c.index
sizes = c['tulos']

fig1, ax1 = plt.subplots(figsize=(10, 10))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=-90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
ykskaks_vali = events_nsu[events_nsu['tilanne'] == '1']
ykskaks_vali_onnistuneet = ykskaks_vali[ykskaks_vali['tulos'] == 'Kärkilyönti']
c = ykskaks_vali_onnistuneet.groupby(by='tyyppi').count().sort_values(by='tulos')

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = c.index
sizes = c['tulos']

fig1, ax1 = plt.subplots(figsize=(10, 10))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=-90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
