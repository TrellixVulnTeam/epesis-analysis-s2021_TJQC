```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

msu = {
    1001: 'KoU',
    1002: 'ViVe',
    1003: 'Tahko',
    1004: 'Manse',
    1005: 'SoJy',
    1006: 'KPL',
    1007: 'JyJu',
    1008: 'IPV',
    1009: 'JoMa',
    1010: 'KaMa',
    1011: 'KeKi',
    1012: 'KiPa',
    1013: 'PattU',
    1014: 'SiiPe',
    1015: 'HP'
}

nsu = {
    2001: 'Fera',
    2002: 'Tahko',
    2003: 'JoMa',
    2004: 'KeKi',
    2005: 'Kirittäret',
    2006: 'Virkiä',
    2007: 'Manse',
    2008: 'MyVe',
    2009: 'PesäYsit',
    2010: 'Pesäkarhut',
    2011: 'SMJ',
    2012: 'SiiPe',
    2013: 'Roihu'
}
```

```python
events_nsu = pd.read_csv('./events_2021_nsu.csv', low_memory=False)
events_msu = pd.read_csv('./events_2021_msu.csv', low_memory=False)
events_msu['tilanne']
```

# Laittomien näppien prosentti joukkueittain

```python
df = pd.DataFrame(columns = ['Joukkue', 'Laiton %', 'Näppien lkmr'])

for team_id in range(1001, 1016):
    evs = events_msu[events_msu['sp'] == team_id]
    evs = evs[(evs['tilanne'] == '1') | (evs['tilanne'] == '1-2')]
    evs = evs[evs['tyyppi'] == 'nappi']
        
    all_napit = len(evs)
    
    all_karkkarit = len(evs[evs['tulos'] == 'Kärkilyönti'])
    
    illegals = len(evs[evs['merkattuLaiton'] == True])
    
    df = df.append({'Joukkue' : msu[team_id], 'Laiton %' : round(illegals / all_napit, 2), 'Näppien lkmr' : all_napit}, ignore_index = True)

df.sort_values(by='Laiton %')
```
