# Lyöntityyppien osuus tilanteittain ja kunkin onnistumisprosentti

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
season = 2021
league = 'nsu'
base = ['1-2', '2', 2]

events = pd.read_csv('./events_{}_{}.csv'.format(season, league), low_memory=False)
events.head()
```

```python
events_by_base = events[events['tilanne'].isin(base)]
by_type = events_by_base['tyyppi'].value_counts()
count = len(events_by_base)

final = pd.DataFrame(columns=['tyyppi', '%', 'kärkipalo', 'haava', 'kärkilyönti'])

for a in by_type.index:
    if(a == 'muu'): break
    all_type = events_by_base = events[events['tyyppi'] == a]
    all_len = len(all_type)
    type_counts = all_type['tulos'].value_counts()
    print(type_counts)
    print(a)
    type_counts = type_counts / all_len
    trio = type_counts['Palo'] + type_counts['Haava'] + type_counts['Kärkilyönti']
    final = final.append({
        'tyyppi': a,
        '%': round(by_type[a] / count, 2),
        'kärkipalo': round(type_counts['Palo'] / trio, 2),
        'haava': round(type_counts['Haava'] / trio, 2),
        'kärkilyönti': round(type_counts['Kärkilyönti'] / trio, 2),
    }, ignore_index=True)

        
```

```python
# Data to plot
ind = {
    'vaakamaila': 0,
    'nappi': 1,
    'pystari/viisto': 2,
    'koppi' : 3,
    'pomppu' : 4,
    'kumura' : 5,
    'varsi': 6,
    'pussari': 7
}

labels = final['tyyppi']
sizes = final['%']
labels_gender = [
    'kärkilyönti', 'kärkipalo', 'haava',
    'kärkilyönti', 'kärkipalo', 'haava',
    'kärkilyönti', 'kärkipalo', 'haava',
    'kärkilyönti', 'kärkipalo', 'haava',
    'kärkilyönti', 'kärkipalo', 'haava',
    'kärkilyönti', 'kärkipalo', 'haava',
    'kärkilyönti', 'kärkipalo', 'haava',
    'kärkilyönti', 'kärkipalo', 'haava',
]
sizes_gender = []

for tyyppi in labels:
    for i in ['kärkilyönti', 'kärkipalo', 'haava']:
        sizes_gender.append(final[i][ind[tyyppi]] * final['%'][ind[tyyppi]])
    
print(sizes_gender)

colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff']
colors_gender = [
    'green','red', 'yellow',
    'green','red', 'yellow',
    'green','red', 'yellow',
    'green','red', 'yellow',
    'green','red', 'yellow',
    'green','red', 'yellow',
    'green','red', 'yellow',
    'green','red', 'yellow',
]
 
# Plot
plt.pie(sizes, labels=labels, colors=colors, startangle=90,frame=True)
plt.pie(sizes_gender,colors=colors_gender,radius=0.75,startangle=90)
centre_circle = plt.Circle((0,0),0.6,color='black', fc='white',linewidth=0)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
 
# plt.axis('equal')
plt.tight_layout()
plt.show()
```
