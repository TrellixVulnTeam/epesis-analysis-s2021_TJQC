%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from IPython.display import display

```python
csv_name = ''
runs_msu = pd.read_csv('../runs/runs_msu_2021.csv'.format(csv_name), low_memory=False)
runs_nsu = pd.read_csv('../runs/runs_nsu_2021.csv'.format(csv_name), low_memory=False)

men_runkosarja = 1630022425
women_runkosarja = 1629849625

runs_msu = runs_msu[runs_msu['lahto'] < 0.6]
runs_nsu = runs_nsu[runs_nsu['lahto'] < 0.6]
runs_msu = runs_msu[runs_msu['unix_time'] < men_runkosarja]
runs_nsu = runs_nsu[runs_nsu['unix_time'] < women_runkosarja]

runs_msu_12 = runs_msu[runs_msu['base'] == 1]
runs_msu_23 = runs_msu[runs_msu['base'] == 2]
runs_nsu_12 = runs_nsu[runs_nsu['base'] == 1]
runs_nsu_23 = runs_nsu[runs_nsu['base'] == 2]
```

```python
# print(runs_msu_12.groupby(by='lukkari').count())

df = runs_msu_12.groupby('lukkari').agg(['count', 'mean'])

df = df[df['id']['count'] > 100]
df['count'] = df['id']['count']
df['time'] = round(df['aika']['mean'], 2)

print('MSU 12')
display(df.sort_values(by='time', ascending=False)[['time', 'count']])

df = runs_msu_23.groupby('lukkari').agg(['count', 'mean'])

df = df[df['id']['count'] > 100]
df['count'] = df['id']['count']
df['time'] = round(df['aika']['mean'], 2)

print('MSU 23')
display(df.sort_values(by='time', ascending=False)[['time', 'count']])

df = runs_nsu_12.groupby('lukkari').agg(['count', 'mean'])

df = df[df['id']['count'] > 60]
df['count'] = df['id']['count']
df['time'] = round(df['aika']['mean'], 2)

print('NSU 12')
display(df.sort_values(by='time', ascending=False)[['time', 'count']])

df = runs_nsu_23.groupby('lukkari').agg(['count', 'mean'])

df = df[df['id']['count'] > 60]
df['count'] = df['id']['count']
df['time'] = round(df['aika']['mean'], 2)

print('NSU 23')
display(df.sort_values(by='time', ascending=False)[['time', 'count']])
```

Huom: Ajat ovat runkosarjan pesänvälin täydet etenemiset joissa lähtö kestää alle 0,6 sekuntia lukkareilta, joilla yli sata (naisilla 80) etenemistä kyseiseltä pesänväliltä
