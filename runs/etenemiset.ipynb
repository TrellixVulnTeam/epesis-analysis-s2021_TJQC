```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csv_name = 'runs_by_week_msu_2021'
runs = pd.read_csv('./{}.csv'.format(csv_name), low_memory=False)
# runs = runs[runs['week'] < 35]
```

```python
runs = runs[runs['base'] == 3]
```

```python
fig, ax = plt.subplots(figsize = (40, 20))

for id in range(1001, 1016):  
#for id in [1006, 1005, 1004, 1002]:  
    team_runs = runs[runs['team'] == id]
    weeks = team_runs.week
    times = team_runs.average_aika
    plt.plot(weeks, times, marker='X', label=id)
    
x = runs.week
y = runs.average_aika
mean_line = plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 5))(np.unique(x)), lw="3", ls="--", color="black", label="Sovitettu suora")
box = mean_line[0]
mean_y = box.get_data()[1]
mean_x = box.get_data()[0]
mean_y0 = mean_y[0]
mean_y1 = mean_y[-1]
mean_x0 = mean_x[0]
mean_x1 = mean_x[-1]
diff = mean_y1 - mean_y0

print(diff)

plt.title('KA-juoksuaika ajan funktiona joukkueittain', fontsize=14)
plt.xlabel('Viikko', fontsize=14)
plt.ylabel('KA juoksuaika', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
